```python
amenities_list = ['Heating', 'Kitchen', 'Air conditioning', 'Hangers', 'Hair dryer', 'Shampoo', 'Iron', 'Laptop friendly workspace',
                  'Hot water', 'Family/kid friendly', 'Coffee maker', 'Extra pillows and blankets', 'Luggage dropoff allowed'
                  'No stairs or steps to enter', 'Lock on bedroom door'] 
```

```python
df['Wifi'] = np.where((df['amenities'].str.contains('Wifi') | df['amenities'].str.contains('Internet')),1,0)
df['TV'] = np.where((df['amenities'].str.contains('TV') | df['amenities'].str.contains('Cable TV')),1,0)
df['Self check-in'] = np.where((df['amenities'].str.contains('Self check-in') | df['amenities'].str.contains('24-hour check-in')),1,0)
df['Free parking'] = np.where((df['amenities'].str.contains('Free street parking') | df['amenities'].str.contains('Free parking on premises')),1,0)
df['Pets allowed'] = np.where((df['amenities'].str.contains('Pets allowed') | df['amenities'].str.contains('Cat(s)') | df['amenities'].str.contains('Dog(s)')),1,0)
df['Pets on property'] = np.where(df['amenities'].str.contains('Pets live on this property'),1,0)

for i in amenities_list:
    df[i] = np.where(df['amenities'].str.contains(i),1,0)
```

**Price & 21 Amenities Graph**

```python
amen_df = df.copy()
amen_df = df[['Heating', 'Kitchen', 'Air conditioning', 'Hangers', 'Hair dryer', 'Shampoo', 'Iron', 'Laptop friendly workspace',
                  'Hot water', 'Family/kid friendly', 'Coffee maker', 'Extra pillows and blankets', 'Luggage dropoff allowed'
                  'No stairs or steps to enter', 'Lock on bedroom door','Wifi','TV','Self check-in','Free parking','Pets allowed','Pets on property','price']]
```

```python
from sklearn.metrics import mean_squared_error, make_scorer
from sklearn.model_selection import GridSearchCV, StratifiedKFold
import lightgbm as lgbm
```

```python
X = amen_df.drop('price', axis='columns').values
y = amen_df['price'].values

const_params = {
    'objective':                      'regression', 
    'metric':                         'rmse',
    'n_jobs':                         -1
}

tuning_parameters = {
    'learning_rate':                  [0.11, 0.115, 0.12], 
    'max_depth':                      [20], 
    'n_estimators':                   [275, 280, 285], 
    'random_state':                   [0]
}

def rmse(y_true, y_pred): return abs(np.sqrt(mean_squared_error(y_true, y_pred)))
scorer = make_scorer(rmse, greater_is_better=False)

grid = GridSearchCV(lgbm.LGBMRegressor(**const_params), 
                    tuning_parameters, 
                    cv=StratifiedKFold(5), 
                    scoring=scorer, 
                    n_jobs=-1)

grid.fit(X, y)

print(f'Best parameters {grid.best_params_}.')
print(f'Best RMSE is {-1*grid.best_score_}.')
```

```python
(pd.Series(grid.best_estimator_.feature_importances_, index=amen_df.drop('price', axis='columns').columns.tolist())
.sort_values(ascending=False)
.plot.bar(color='crimson', figsize=(16, 5))).figure.savefig("prices&amenities");
```

**Price & 21 Amenities Linear Regression**

```python
import statsmodels.api as sm
import seaborn as sns

sns.set(rc={'figure.figsize':(15,8)})
```

```python
X = amen_df.drop('price', axis='columns')
y = amen_df['price']

model_price = sm.OLS(y, X).fit()
predictions = model_price.predict(X) 
model_price.summary()
```

```python
coeff = model_price.params
data = [[x, y] for x,y in zip(coeff.values, coeff.index.tolist())]
coeff_df = pd.DataFrame(data, columns=['values','amenities'])
coeff_df = coeff_df.sort_values(by=['values'],ascending=False)
coeff_df
```

```python
ax = sns.barplot(x = 'amenities', y = 'values', data = coeff_df, palette = sns.color_palette('Reds_d',21))
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
```

```python
ax.figure.savefig("prices&amenities.png")
```
